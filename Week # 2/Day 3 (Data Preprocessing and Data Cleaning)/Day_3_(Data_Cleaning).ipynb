# Steps for Data Preprocessing

1. Renaming the columsn
2. Deleting redundant columns
3. dropping duplicated
4. Removing the NaN values
5. Cleaning individual columns

6. Check for some more transfomration




# 1. Renaming the columns


```python
dataset.columns
```

```python
dataset.columns = (
    dataset.columns.str.strip()
              .str.lower()
              .str.replace(r"[ -]+", "_", regex=True)
              .str.replace(r"[^\w_]", "", regex=True)
)
```

```python
dataset.columns
```

```python
dataset.info()
```

# 2. Deleting redundant columns

```python
columns_to_drop = ["license", "house_rules", "calculated_host_listings_count", "reviews_per_month", "review_rate_number", "id", "availability_365", "last_review"]
```

```python
dataset[columns_to_drop]
```

```python
dataset.drop(columns=columns_to_drop, inplace = True)
```

```python
dataset
```

# 3. Deleting duplicates

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.duplicated().sum()
```

# 4. Removing the NaN values

```python
dataset.isnull()
```

```python
dataset.isnull().value_counts()
```

```python
dataset.isnull().sum()
```

```python
(dataset.isnull().sum()/dataset.shape[0])*100
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```
